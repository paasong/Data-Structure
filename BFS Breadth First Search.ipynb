## BFS 구현하기

Graph 파이썬으로 구현하기 <br>
dict을 이용하여 구현함

Key : Value <br>
A : B C <br>
B : A D <br>
C : A G H I <br>
D : B E F <br>
E : D <br>
F : D <br>
G : C <br>
H : C <br>
I : C J <br>
J : I <br>

```python
# 그래프 생성하기
graph = dict()

graph['A'] = ["B", "C"]
graph['B'] = ["A", "D"]
graph['C'] = ["A", "G", "H", "I"]
graph['D'] = ["B", "E", "F"]
graph["E"] = ["D"]
graph["F"] = ["D"]
graph["G"] = ["C"]
graph["H"] = ["C"]
graph["I"] = ["C", "J"]
graph["J"] = ["I"]
```

### BFS Parameter <br>
Need_visit Queue: 노드마다 방문해야하는 노드<br>
Visited Queue: 방문한 노드

```python
def bfs(graph, start_node):
    visited = list()
    need_visit = list()
    need_visit.append(start_node)
    
    
    while need_visit:
        node = need_visit.pop(0)
        if node not in visited:
            visited.append(node)
            need_visit.extend(graph[node])

    return visited
bfs(graph, 'A')
```
